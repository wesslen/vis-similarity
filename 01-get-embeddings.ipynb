```python
import numpy as np
from gensim.corpora import Dictionary
from gensim.models.tfidfmodel import TfidfModel
from gensim.matutils import sparse2full
from sklearn.decomposition import TruncatedSVD

#df = df[df.Conference.isin(['InfoVis','VAST','Vis','SciVis','DECISIVe','TVCG'])]

docs = [_ for _ in df['Abstract'] if _ != "nan"]

import spacy

nlp = spacy.load("en_core_web_lg", disable=("parser", "tagger", "ner"))
```

```python
## see https://github.com/PrincetonML/SIF

def compute_pc(X,npc=1):
    """
    Compute the principal components. DO NOT MAKE THE DATA ZERO MEAN!
    :param X: X[i,:] is a data point
    :param npc: number of principal components to remove
    :return: component_[i,:] is the i-th pc
    """
    svd = TruncatedSVD(n_components=npc, n_iter=7, random_state=0)
    svd.fit(X)
    return svd.components_

def remove_pc(X, npc=1):
    """
    Remove the projection on the principal components
    :param X: X[i,:] is a data point
    :param npc: number of principal components to remove
    :return: XX[i, :] is the data point after removing its projection
    """
    pc = compute_pc(X, npc)
    if npc==1:
        XX = X - X.dot(pc.transpose()) * pc
    else:
        XX = X - X.dot(pc.transpose()).dot(pc)
    return XX
```

```python
def keep_token(t):
    return (t.is_alpha and 
            not (t.is_space or t.is_punct or 
                 t.is_stop or t.like_num))

def lemmatize_doc(doc):
    return [t.lemma_ for t in doc if keep_token(t)]

def get_vectors(docs, svd_reduction = True):

    docs2 = [lemmatize_doc(nlp(doc)) for doc in docs]

    docs_dict = Dictionary(docs2)
    docs_dict.filter_extremes(no_below=10, no_above=0.2)
    docs_dict.compactify()

    docs_corpus = [docs_dict.doc2bow(doc) for doc in docs2]
    model_tfidf = TfidfModel(docs_corpus, id2word=docs_dict)
    docs_tfidf  = model_tfidf[docs_corpus]
    docs_vecs   = np.vstack([sparse2full(c, len(docs_dict)) for c in docs_tfidf])
    tfidf_emb_vecs = np.vstack([nlp(docs_dict[i]).vector for i in range(len(docs_dict))])

    docs_emb = np.dot(docs_vecs, tfidf_emb_vecs) 
    
    if svd_reduction:
        docs_emb = remove_pc(np.transpose(docs_emb),npc=1)
        docs_emb = np.transpose(docs_emb)

    return docs_emb

def cos_sim(v1,v2):
    cs = (np.dot(v1,v2)) / ((np.linalg.norm(v1)) * (np.linalg.norm(v2)))
    return cs

def get_sim(q, docs_emb):
    
    sims = []
    for i in range(len(docs_emb)):
        s = cos_sim(docs_emb[q],docs_emb[i])
        sims.append(s)
    
    zipped = zip(range(len(docs_emb)),sims)
    d = dict(zipped)
    
    sorted_x = sorted(d.items(), key=lambda kv: kv[1], reverse=True)
    sim_df = pd.DataFrame(sorted_x, columns =['Index','Similarity']) # modify to Index
    sim_df = sim_df.set_index('Index')
    
    return sim_df

def get_results(i, docs_emb, df):

    f = get_sim(i, docs_emb)
    ret = {"Similarity": f.Similarity,
        "Title": df.Title[f.index],
        #"Authors": df.AuthorNamesDeduped[f.index],
        "Authors": df.Authors[f.index], # beware -- if you old vispapers, you'll need to change
        "Conference": df.Conference[f.index],
        "Year": df.Year[f.index],
        "Keywords": df.Keywords[f.index]
    }
    
    
    ret = pd.DataFrame(ret)
    
    ret = ret.drop(i)

    return ret
```
